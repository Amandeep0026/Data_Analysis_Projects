```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("E:\COURSE_DATA_SCIENCE\Projects\Project_2\Onyx Data DataDNA Datatset Challenge - Spotify Most Streamed Songs 2023 Dataset - October 2023.csv")
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Alternatively, you could fill missing values
df['in_shazam_charts'] = df['in_shazam_charts'].fillna(0)
```

```python
# Alternatively, you could fill missing values
df['key'] = df['key'].fillna('None')
```

```python
# Display the data types of the columns
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics
df.describe().transpose()
```

```python
# Correlation matrix
corr_matrix = df.corr(method="pearson", numeric_only=True)

# Heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_matrix, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
# Bar plot of tracks released per year
plt.figure(figsize=(12, 6))
sns.countplot(x='released_year', data=df, palette="magma")
plt.xticks(rotation=90)
plt.title('Number of songs per year')
plt.xlabel('Released Year')
plt.ylabel('Count of Tracks')
plt.show()
```

```python
# Count of tracks per artist
top_artists = df['artist(s)_name'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_artists.index, y=top_artists.values, palette="dark:salmon_r")
plt.xticks(rotation=90)
plt.title('Top 10 Artists by Track')
plt.xlabel('Artist Name')
plt.ylabel('Count of Tracks')
plt.show()
```

```python
# Pairplot for danceability, energy, and valence
sns.pairplot(df[['danceability_%', 'energy_%', 'valence_%']])
plt.show()
```

```python

# Scatter plot for danceability vs energy
plt.figure(figsize=(12, 6))
sns.scatterplot(x='danceability_%', y='energy_%', data=df, hue='mode')
plt.title('Danceability vs Energy')
plt.show()
plt.figure(figsize=(12, 6))
sns.regplot(data =df, y = "danceability_%", x = "energy_%", color = "c").set(title = "Danceability vs Energy Correlation")
```

```python
# Bar plot of tracks in different playlists
playlists = ['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']
playlist_counts = df[playlists].sum().sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=playlist_counts.index, y=playlist_counts.values)
plt.title('Number of Tracks in Different Playlists')
plt.show()
```
